```python
stemmer = Stemmer()
lemmatizer = Lemmatizer()
stopword_dict = {sw:0 for sw in stopwords_list()}
description_df

def normalize_description(df,feature):
    i=0
    for row in df[feature]:
        tokenized_sentence = word_tokenize(row)
        new_sentence = ""
        for word in tokenized_sentence:
            if word in  [':','؟','.',',','«','»','(',')','،','[',']','{','}','-','','؛', '\r', '\n']:
                continue
            if word in stopword_dict:
                continue
            word = stemmer.stem(word)
            lemmitized_word = lemmatizer.lemmatize(word)
            new_sentence+=(lemmitized_word) 
        df[feature][i] = new_sentence
#         print(new_sentence)
        i += 1
        if i == 1000:
            break
    return df
# print(description_df['description'][0])
description_df = normalize_description(description_df, "description")
description_df['description'][0]
```
